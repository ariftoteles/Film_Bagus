    - Widodo sangat menyukai film drama komedi, salah satunya bertajuk _**Being Flynn (2012)**_.

```python
import numpy as np
import pandas as pd
```

```python
rating = pd.read_csv('ratings.csv')
movie = pd.read_csv('movies.csv')
rating
```

### Cleaning Data

```python
rating.isnull().sum()
```

```python
rating.info()
```

```python
rating_piv = rating.pivot_table(index='userId',columns='movieId',values='rating')
rating_piv = rating_piv.fillna(0)
rating_piv
```

```python
ratingCorr = rating_piv.corr()
ratingCorr
```

## Rekomendasi untuk Widodo

- Widodo sangat menyukai film drama komedi, salah satunya bertajuk _Being Flynn (2012)_

- Karena Widodo sangat menyukai film tersebut, kemungkinan besar Widodo memberi rating 5

```python
## cek movie ID Being Flynn
widodo = movie['movieId'][movie['title']=='Being Flynn (2012)'].values[0]
widodo
```

```python
## movie ID for film Being Flynn (2012)
skor = ratingCorr[widodo] * 5
skor = skor.sort_values(ascending=False)
skor = pd.DataFrame(skor).reset_index()
skor.head()
```

```python
## Membuat Movie Being Flynn di idex paling atas
df2 = skor[skor['movieId']==widodo]
skor_new = pd.concat([df2,skor],axis=0)
skor_new = skor_new.drop_duplicates(subset='movieId').reset_index()
skor_new = skor_new.drop(columns='index')
skor_new 
```

```python
rekWidodo = pd.merge(skor_new,movie,left_on='movieId',right_on='movieId')
rekWidodo = rekWidodo.drop(columns=95816)
rekWidodo
```
